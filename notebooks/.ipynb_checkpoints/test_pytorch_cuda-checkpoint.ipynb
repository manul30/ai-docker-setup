# PyTorch CUDA Test & MobileNetV3 Inference
Test CUDA availability, performance benchmarks, and real model inference with MobileNetV3

```python
# Import libraries
import torch
import torchvision
import time
import numpy as np
from torchvision import models, transforms
from PIL import Image
import matplotlib.pyplot as plt

print(f"PyTorch version: {torch.__version__}")
print(f"Torchvision version: {torchvision.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA version: {torch.version.cuda}")
    print(f"cuDNN version: {torch.backends.cudnn.version()}")
    print(f"GPU count: {torch.cuda.device_count()}")
    print(f"Current GPU: {torch.cuda.current_device()}")
    print(f"GPU name: {torch.cuda.get_device_name(0)}")
    print(f"GPU memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
```

## CUDA Performance Benchmarks
Compare CPU vs GPU performance with matrix multiplications

```python
# Performance benchmark: Matrix multiplication
size = 4096
repeats = 5

print(f"\nBenchmarking {size}x{size} matrix multiplication ({repeats} runs)\n" + "="*60)

# CPU benchmark
print("\n[CPU Benchmark]")
cpu_times = []
for i in range(repeats):
    a = torch.randn(size, size)
    b = torch.randn(size, size)
    start = time.time()
    c = torch.matmul(a, b)
    cpu_time = time.time() - start
    cpu_times.append(cpu_time)
    print(f"  Run {i+1}: {cpu_time:.4f} seconds")

cpu_avg = np.mean(cpu_times)
print(f"  Average: {cpu_avg:.4f} seconds")

# GPU benchmark
if torch.cuda.is_available():
    print("\n[GPU Benchmark]")
    gpu_times = []
    device = torch.device('cuda')
    
    # Warmup
    a_gpu = torch.randn(size, size, device=device)
    b_gpu = torch.randn(size, size, device=device)
    _ = torch.matmul(a_gpu, b_gpu)
    torch.cuda.synchronize()
    
    for i in range(repeats):
        a_gpu = torch.randn(size, size, device=device)
        b_gpu = torch.randn(size, size, device=device)
        torch.cuda.synchronize()
        start = time.time()
        c_gpu = torch.matmul(a_gpu, b_gpu)
        torch.cuda.synchronize()
        gpu_time = time.time() - start
        gpu_times.append(gpu_time)
        print(f"  Run {i+1}: {gpu_time:.4f} seconds")
    
    gpu_avg = np.mean(gpu_times)
    print(f"  Average: {gpu_avg:.4f} seconds")
    print(f"\n  ⚡ GPU Speedup: {cpu_avg/gpu_avg:.2f}x faster than CPU")
    print(f"  GPU memory allocated: {torch.cuda.memory_allocated()/1e9:.2f} GB")
    print(f"  GPU memory reserved: {torch.cuda.memory_reserved()/1e9:.2f} GB")
```

## Mixed Precision (FP16) Benchmark
Test automatic mixed precision training performance

```python
if torch.cuda.is_available():
    print("\nMixed Precision (FP16) Benchmark\n" + "="*60)
    device = torch.device('cuda')
    
    # Create a simple model
    model = torch.nn.Sequential(
        torch.nn.Linear(1024, 2048),
        torch.nn.ReLU(),
        torch.nn.Linear(2048, 1024),
        torch.nn.ReLU(),
        torch.nn.Linear(1024, 10)
    ).to(device)
    
    x = torch.randn(128, 1024, device=device)
    
    # FP32 (default)
    print("\n[FP32 Forward Pass]")
    fp32_times = []
    for i in range(10):
        torch.cuda.synchronize()
        start = time.time()
        output = model(x)
        torch.cuda.synchronize()
        fp32_times.append(time.time() - start)
    fp32_avg = np.mean(fp32_times[3:])  # Skip first 3 warmup runs
    print(f"  Average time: {fp32_avg*1000:.2f} ms")
    
    # FP16 with autocast
    print("\n[FP16 Forward Pass (Automatic Mixed Precision)]")
    fp16_times = []
    for i in range(10):
        torch.cuda.synchronize()
        start = time.time()
        with torch.cuda.amp.autocast():
            output = model(x)
        torch.cuda.synchronize()
        fp16_times.append(time.time() - start)
    fp16_avg = np.mean(fp16_times[3:])  # Skip first 3 warmup runs
    print(f"  Average time: {fp16_avg*1000:.2f} ms")
    print(f"\n  ⚡ FP16 Speedup: {fp32_avg/fp16_avg:.2f}x faster than FP32")
```

## MobileNetV3 Object Detection Inference
Load MobileNetV3 and perform inference on sample images

```python
# Load MobileNetV3 (classification model)
print("Loading MobileNetV3 Large model...")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load pre-trained MobileNetV3
model = models.mobilenet_v3_large(weights=models.MobileNet_V3_Large_Weights.DEFAULT)
model = model.to(device)
model.eval()

print(f"✓ Model loaded on {device}")
print(f"  Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"  Model size: {sum(p.numel() * p.element_size() for p in model.parameters()) / 1e6:.2f} MB")

# Define preprocessing transforms
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

print("\n✓ Preprocessing pipeline ready")
```

```python
# Create a sample image for testing
import urllib.request

# Download ImageNet class labels
print("Downloading ImageNet class labels...")
LABELS_URL = "https://raw.githubusercontent.com/anishathalye/imagenet-simple-labels/master/imagenet-simple-labels.json"
import json

try:
    with urllib.request.urlopen(LABELS_URL) as url:
        labels = json.loads(url.read().decode())
    print(f"✓ Loaded {len(labels)} class labels")
except Exception as e:
    print(f"Warning: Could not load labels: {e}")
    labels = [f"class_{i}" for i in range(1000)]

# Create a synthetic test image (or you can load a real image)
print("\nCreating synthetic test image...")
test_image = Image.new('RGB', (224, 224), color='blue')
# Add some pattern to make it more interesting
for i in range(0, 224, 20):
    for j in range(0, 224, 20):
        if (i + j) % 40 == 0:
            test_image.putpixel((i, j), (255, 0, 0))

# Display the test image
plt.figure(figsize=(5, 5))
plt.imshow(test_image)
plt.title('Test Image')
plt.axis('off')
plt.show()

print("✓ Test image created")
```

```python
# Perform inference
print("\nPerforming MobileNetV3 Inference\n" + "="*60)

# Preprocess the image
input_tensor = preprocess(test_image)
input_batch = input_tensor.unsqueeze(0).to(device)  # Create a mini-batch

print(f"Input tensor shape: {input_batch.shape}")
print(f"Input tensor device: {input_batch.device}")

# Warmup run
with torch.no_grad():
    _ = model(input_batch)

if torch.cuda.is_available():
    torch.cuda.synchronize()

# Timed inference
inference_times = []
num_runs = 100

print(f"\nRunning {num_runs} inference iterations...")
with torch.no_grad():
    for i in range(num_runs):
        if torch.cuda.is_available():
            torch.cuda.synchronize()
        start = time.time()
        output = model(input_batch)
        if torch.cuda.is_available():
            torch.cuda.synchronize()
        inference_times.append(time.time() - start)

# Get predictions
probabilities = torch.nn.functional.softmax(output[0], dim=0)
top5_prob, top5_catid = torch.topk(probabilities, 5)

print("\n" + "="*60)
print("Inference Results:")
print("="*60)
print(f"Average inference time: {np.mean(inference_times)*1000:.2f} ms")
print(f"Min inference time: {np.min(inference_times)*1000:.2f} ms")
print(f"Max inference time: {np.max(inference_times)*1000:.2f} ms")
print(f"Throughput: {1/np.mean(inference_times):.2f} images/second")

print("\nTop 5 Predictions:")
for i in range(top5_prob.size(0)):
    print(f"  {i+1}. {labels[top5_catid[i]]}: {top5_prob[i].item()*100:.2f}%")

if torch.cuda.is_available():
    print(f"\nGPU Memory Usage:")
    print(f"  Allocated: {torch.cuda.memory_allocated()/1e6:.2f} MB")
    print(f"  Reserved: {torch.cuda.memory_reserved()/1e6:.2f} MB")
    print(f"  Max allocated: {torch.cuda.max_memory_allocated()/1e6:.2f} MB")
```

## Batch Inference Performance
Test inference with different batch sizes

```python
print("\nBatch Inference Performance Test\n" + "="*60)

batch_sizes = [1, 4, 8, 16, 32]
batch_results = {}

for batch_size in batch_sizes:
    print(f"\nTesting batch size: {batch_size}")
    
    # Create batch
    batch = torch.randn(batch_size, 3, 224, 224, device=device)
    
    # Warmup
    with torch.no_grad():
        _ = model(batch)
    
    if torch.cuda.is_available():
        torch.cuda.synchronize()
    
    # Benchmark
    times = []
    with torch.no_grad():
        for _ in range(50):
            if torch.cuda.is_available():
                torch.cuda.synchronize()
            start = time.time()
            _ = model(batch)
            if torch.cuda.is_available():
                torch.cuda.synchronize()
            times.append(time.time() - start)
    
    avg_time = np.mean(times)
    throughput = batch_size / avg_time
    batch_results[batch_size] = {
        'avg_time': avg_time,
        'throughput': throughput,
        'time_per_image': avg_time / batch_size
    }
    
    print(f"  Avg time: {avg_time*1000:.2f} ms")
    print(f"  Time per image: {avg_time/batch_size*1000:.2f} ms")
    print(f"  Throughput: {throughput:.2f} images/second")

# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(list(batch_results.keys()), 
         [v['time_per_image']*1000 for v in batch_results.values()], 
         marker='o', linewidth=2, markersize=8)
plt.xlabel('Batch Size')
plt.ylabel('Time per Image (ms)')
plt.title('Inference Time per Image vs Batch Size')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(list(batch_results.keys()), 
         [v['throughput'] for v in batch_results.values()], 
         marker='s', linewidth=2, markersize=8, color='orange')
plt.xlabel('Batch Size')
plt.ylabel('Throughput (images/second)')
plt.title('Throughput vs Batch Size')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n✓ Batch inference benchmark complete!")
```
